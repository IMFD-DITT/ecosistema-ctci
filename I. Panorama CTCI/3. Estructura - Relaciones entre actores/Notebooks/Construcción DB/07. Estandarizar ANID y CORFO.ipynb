## Añadir año a datos de CORFO Ley I+D

Hay datos de CORFO que no tienen año.

```python
df_corfo.loc[df_corfo["año_adjudicacion"].isnull()]
```

Y vemos que todos estos datos son de Ley I+D.

```python
df_corfo.loc[df_corfo["año_adjudicacion"].isnull(), "instrumento"].unique()
```

Intentaremos rellenar estos datos. Para eso, usaremos el excel descargado desde [DataInnovacion](https://datainnovacion.cl/buscador-proyectos)

```python
df_external = pd.read_excel(os.path.join(corfo_data_dir, "DataInnovacion - InnovaChile Corfo.xlsx"), skiprows=1) # first row is empty
df_external
```

```python
df_corfo
```

```python
df_corfo_merged = df_corfo.merge(df_external, left_on="codigo", right_on="Código Proyecto", how="left")
df_corfo_merged
```

```python
df_corfo_merged.loc[df_corfo_merged["año_adjudicacion"].isnull(), "Año Adjudicación"].unique()
```

```python
df_corfo_merged["año_final"] = df_corfo_merged["año_adjudicacion"].combine_first(df_corfo_merged["Año Adjudicación"])
df_corfo_merged[df_corfo_merged["año_final"].isnull()]
```

Hay sólo 16 casos donde no hay año. Me da la impresión que son datos de 2023, pero no estoy seguro al respecto.

```python
df_corfo_merged[(df_corfo_merged["año_final"] != df_corfo_merged["año_adjudicacion"]) &
                (~df_corfo_merged["año_adjudicacion"].isnull())]
```

Todo bien. Amononaremos el dataframe, borrando columnas innecesarias.

```python
df_corfo_merged.columns
```

```python
rows_to_delete = ['Código Proyecto', 'Foco Apoyo', 'Tipo Intervención', 'Instrumento',
                  'Instrumento Homologado', 'Estado Data', 'Tipo Persona',
                  'Rut Beneficiario', 'Beneficiario', 'Título', 'Objetivo',
                  'Año Adjudicación', 'Financiamiento Innova', 'Aprobado Privado',
                  'Aprobado Privado Pecuniario', 'Monto Certificado Ley',
                  'Tipo Innovación', 'Mercado Objetivo', 'Criterio Mujer', 'Género',
                  'Sostenible', 'ODS principal', 'Meta principal', 'Economía Circular',
                  'Modelo de Circularidad', 'Región', 'Tramo Ventas',
                  'Inicio Actividad Económica', 'Sector Económico',
                  'Patron principal asociado', 'Tipo proyecto', 'R principal',
                  'Estrategia R Principal', 'Ley REP', 'Ley REP (Sí/No)', 'ERNC',
                  'Tendencia Final', "año_adjudicacion"]
rows_order = ['id_fila', 'id', 'año_adjudicacion', 'codigo', 'instrumento',
            'tipo_relacion', 'titulo_del_proyecto', 'rutdv', 'nombre_razon_social',
            'tipo_ctr_sii', 'rubro_sii', 'aprobado_corfo', 'aprobado_privado_total',
            'aprobado_privado_pecuniario', 'monto_consolidado_ley', 'rut_valido',
            'nombre_normalizado', 'rut_dv', 'nombre_canonico', 'empresa_tipo',
            'empresa_subtipo', 'provincia', 'comuna', 'tipo_institucion_final']
df_corfo = df_corfo_merged.drop(columns=rows_to_delete) \
                          .rename(columns={"año_final": "año_adjudicacion"}) \
                          [rows_order]
```
