```python
import os
import pandas as pd
```

## Cargar datos

```python
data_dir = "/content/drive/Shareddrives/Estudio Ecosistema CTCI/PT2_Sistematización fuentes de información/Análisis/ANID_CORFO/Datos/ANID"
results_dir = "/content/drive/Shareddrives/Estudio Ecosistema CTCI/PT2_Sistematización fuentes de información/Análisis/ANID_CORFO/Resultados/ANID"
```

```python
df_institutions = pd.read_excel(os.path.join(results_dir, "data_instituciones_anid.xlsx"))
df_institutions["rut"] = df_institutions["rut"].astype(str)
df_institutions
```

```python
df_anid = pd.read_excel(os.path.join(results_dir, "data_anid_ruts_imputados.xlsx"))
df_anid
```

## Juntar dataframes

```python
df_merged = df_anid.merge(df_institutions, left_on="rut_normalizado", right_on="rut_dv", how="left")
df_merged
```

## Análisis

Revisaremos los casos donde no hay `tipo_institucion_final`

```python
df_merged[df_merged["tipo_institucion_final"].isnull()]
```

Las dividiremos en instituciones chilenas e internacionales.

### Internacionales

```python
df_merged.loc[(df_merged["tipo_institucion_final"].isnull())
              & (df_merged["PAISINSTITUCION"] != "CHILE")]
```

Son 2237 filas. A todas ellas, en la columna `tipo_institucion_final` les pondremos el valor _Internacional_.

```python
df_merged.loc[(df_merged["tipo_institucion_final"].isnull())
              & (df_merged["PAISINSTITUCION"] != "CHILE"),
              "tipo_institucion_final"] = "Internacional"
df_merged.loc[df_merged["tipo_institucion_final"] == "Internacional"]
```

### Nacionales

```python
df_merged.loc[(df_merged["tipo_institucion_final"].isnull())
              & (df_merged["PAISINSTITUCION"] == "CHILE")]
```

730 filas nacionales. Analicemos cuántas tienen RUT válido.

```python
df_merged.loc[(df_merged["tipo_institucion_final"].isnull())
              & (df_merged["PAISINSTITUCION"] == "CHILE")
              & (df_merged["rut_valido"])]
```

Son sólo 68 filas.

### Arreglar nombres

En df_merged, hay 3 columnas relacionadas al nombre de la institución:
- `INSTITUCION1ERNIVEL`: nombre original que venía en los datos de ANID.
- `nombre_normalizado`: normalización del campo `INSTITUCION1ERNIVEL` aplicando ciertas transformaciones
- `nombre`: nombre extraído desde fuente externa (PortalChile o Registros19862)

Esta última columna la renombraremos a `nombre_canonico`.

```python
df_merged = df_merged.rename(columns={"nombre": "nombre_canonico"})
df_merged
```

Veamos las instituciones que no tienen `nombre_normalizado`

```python
df_merged[df_merged["nombre_normalizado"].isnull()]
```

Todas tienen `nombre_normalizado`. Veamos ahora cuántas carecen de `nombre_canonico`.

```python
df_merged[df_merged["nombre_canonico"].isnull()]
```

Para todos estos casos, pondremos `nombre_canonico` igual a `nombre_normalizado`.

```python
idx_null = df_merged["nombre_canonico"].isnull()
df_merged.loc[idx_null, "nombre_canonico"] = df_merged.loc[idx_null, "nombre_normalizado"]
df_merged.loc[idx_null]
```

### Un nombre por RUT

```python
grouped_rut = df_merged.groupby(["rut_dv"], as_index=False).agg({"nombre_canonico": set})
grouped_rut[grouped_rut["nombre_canonico"].str.len() != 1]
```

Todos los RUTs sólo tienen un nombre canónico

### Un RUT por nombre

```python
grouped_name = df_merged.groupby(["nombre_canonico"], as_index=False).agg({"rut_dv": set})
grouped_name[grouped_name["rut_dv"].str.len() != 1]
```

Hay dos casos. Veamos DMC SPA primero.

```python
df_merged[df_merged["nombre_canonico"] == "DMC SPA"]
```

Al parecer son instituciones distintas. Les cambiaremos el nombre para desambiguarlas.

```python
df_merged.loc[df_merged["rut_dv"] == "76329676-8", "nombre_canonico"] = "DISTRIBUCION MONTAJE Y COMISIONES LEMUSSE Y COMPANIA LIMITADA"
df_merged.loc[df_merged["rut_dv"] == "76253844-k", "nombre_canonico"] = "DMC SPA"
df_merged.loc[df_merged["rut_dv"].isin(["76329676-8", "76253844-k"])]
```

Veamos ahora el Servicio Nacional de Pesca

```python
df_merged.loc[df_merged["nombre_canonico"] == "SERVICIO NACIONAL DE PESCA"]
```

En este caso, el Servicio Nacional de Pesca tiene dos RUTs asociados: 60701005-6 y 60701004-8. Si bien ambos son RUTs válidos y bien formados, sólo el primer RUT aparece en PortalChile y Registros19862. Además, revisando los datos en internet, vemos que el RUT de Sernapesca general (no direcciones regionales) es 60701002-1. Veamos si tenemos ese RUT registrado:

```python
df_merged.loc[df_merged["rut_dv"] == "60701002-1"]
```

Ese RUT tiene asociado el nombre (casi idéntico, pero distinto al fin y al cabo) de _SERV NACIONAL DE PESCA_. Lo que haremos será asignar a todos los SERVICIO NACIONAL DE PESCA los datos de SERV NACIONAL DE PESCA.

```python
idx_sernapesca = df_merged["nombre_canonico"] == "SERVICIO NACIONAL DE PESCA"
df_merged.loc[idx_sernapesca, ["rut", "rut_dv", "nombre_canonico", "empresa_tipo", "empresa_subtipo", "provincia", "comuna", "tipo_institucion_final"]] = \
              ("60701002",	"60701002-1",	"SERV NACIONAL DE PESCA", "INSTITUCIONES FISCALES", "ORG. ADMINISTRACION PUBLICA", "VALPARAÍSO", "VALPARAISO", 	"Estado")
df_merged.loc[idx_sernapesca]
```

Voldemos a revisar que sólo haya un RUT por cada nombre.

```python
grouped_name = df_merged.groupby(["nombre_canonico"], as_index=False).agg({"rut_dv": set})
grouped_name[grouped_name["rut_dv"].str.len() != 1]
```
